```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen, Request
import requests
from lxml import html, etree
import pandas as pd
from time import time
```

```python
response=requests.get('https://www.jornadaperfecta.com')
response.status_code
soup=bs(requests.get('https://www.jornadaperfecta.com/').text, 'lxml')
equipos=[]
for elem in soup.find("div",{"class":"listado-escudos col-9-12 mobile-col-1-1"}).findAll("a"):
    equipos.append(elem['href'])
equipos
```

```python
def getListplayers(team):
    soup = bs(requests.get(team).text, 'lxml')
    players=[]
    [players.append(elem['href']) for elem in soup.find("div",{"class":"ficha-equipo-tabla"}).findAll("a",itemprop="url")]        
    return(players)
players=list(map(getListplayers, equipos))
players
```

```python
def getinfoplayers(player):
    #Now, we are going to web scraping each link to get the neccesary values to create the starting dataframe.
    row_player=[] #We create empty list for each player
    historical_points=[] #We create empty list for historial points from each player
    soup1 = bs(requests.get(player).text, 'lxml') ##We request to get the info from player.
    row_player.append(soup1.find({"div","h3"},itemprop="name").get_text()) #We get the name of the player.
    row_player.append(soup1.find("div", { "class" : "jugador-demarcacion" }).find('div').get_text().splitlines()[1]) #We get the position of the player.
    row_player.append(soup1.find("div",{"class":"jugador-datos-cifra"}).get_text()) #We get the points from the player.
    row_player.append(soup1.find("div",{"class":"jugador-datos-puntos"}).text.split(' ')[2]) #We get the home points from the player.
    row_player.append(soup1.find("div",{"class":"jugador-datos-puntos"}).text.split(' ')[4]) #We get the away points from the player.
    #soup1.find("div",{"class":"jugador-datos-partidos"}).get_text().split('\n') #Let's get info from number of matches.
    row_player.append(soup1.find("div",{"class":"jugador-datos-partidos"}).get_text().split('\n')[1]) #Number of total matches.
    row_player.append(soup1.find("div",{"class":"jugador-datos-partidos"}).get_text().split('\n')[6]) #Number of home matches.
    row_player.append(soup1.find("div",{"class":"jugador-datos-partidos"}).get_text().split('\n')[10]) #Number of away matches.
    aux_list=[elem for i in soup1.findAll("div",{"class":"jugador-historico-fila"}) for j in i for elem in j]
    row_player.extend(aux_list[2:5])
    row_player.extend(aux_list[8:11])
    row_player.extend(aux_list[14:17])
    row_player.append(soup1.find("a", itemprop="worksFor").find("img")['title']) #Team
    row_player.append(soup1.find("div",{"class":"jugador-datos-precio"}).find("div",itemprop="netWorth").text) #Now, we concatenate the data.
    
    #Now, we replace the missings value for 0.
    
    row_player = ['0' if x==' ' else x for x in row_player]
    return row_player
```

```python
time_inicial=time()
info_to_df=[getinfoplayers(player) for team in players for player in team]
time_final=time()
print('El tiempo fue', time_final-time_inicial)
```
